```python
import pickle
import matplotlib.pyplot as plt
```

# characters

```python
a_unigram = {}
a_bigram = {}
a_trigram = {}
a_quadgram = {}
```

```python
with open('3a_unigram.pickle', 'rb') as handle:
    a_unigram = pickle.load(handle)
with open('3a_bigram.pickle', 'rb') as handle:
    a_bigram = pickle.load(handle)
with open('3a_trigram.pickle', 'rb') as handle:
    a_trigram = pickle.load(handle)
with open('3a_quadgram.pickle', 'rb') as handle:
    a_quadgram = pickle.load(handle)
```

```python
def plot_graph(dict_name,what):
    x= []
    y = []
    cnt=1
    for k,v in dict_name.items():
        x.append(cnt)
        y.append(v)
        cnt+=1
    plt.plot(x, y)
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.title(what)
    plt.show()   
```

```python
plot_graph(a_unigram,'unigram_characters')
```

```python
plot_graph(a_bigram,'bigram_characters')
```

```python
plot_graph(a_trigram,'trigram_characters')
```

```python
plot_graph(a_quadgram,'quadgram_characters')
```
